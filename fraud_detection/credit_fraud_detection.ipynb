# Import Libraries and Load Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

# Exloratory Data Analysis

```python
df.info()
df['Class'].value_counts()
```

```python
sns.histplot(df['Amount'], bins=100, kde=True)
```

```python
sns.heatmap(df.corr(),cmap='coolwarm')
```

# Preprocessing

```python
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])
df.drop('Time', axis=1, inplace=True)
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

# Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42) # stratify so testset represent both classes
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

# Evaluate the model

```python
y_pred = rf_classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC AUC Score: ", roc_auc_score(y_test, rf_classifier.predict_proba(X_test)[:,1]))
```

# Visualization

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')
fpr, tpr, _ = roc_curve(y_test, rf_classifier.predict_proba(X_test)[:,1])
plt.plot(fpr, tpr)
```

# Save the Model

```python
import joblib
joblib.dump(rf_classifier, 'rf_fraud_model.pkl')
```
